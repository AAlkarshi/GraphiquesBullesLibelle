```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("camembert.csv", header=0, sep=",", skipinitialspace=True)

categories = df['Categories']
votes = df['Votes']

# Crer le graphique  bulles
plt.figure(figsize=(12, 8))
scatter = plt.scatter(categories, votes, s=votes*10, alpha=0.5)

# Ajouter des libells pour chaque bulle
for i in range(len(categories)):
    plt.text(categories[i], votes[i], categories[i], 
             horizontalalignment='center', 
             verticalalignment='bottom', 
             fontsize=10, 
             color='black')

plt.title('Rpartition des votes par catgorie')
plt.ylabel('Votes')
plt.xticks(rotation=90)  # Rotation des labels pour une meilleure lisibilit
plt.grid(True)
plt.show()
```
